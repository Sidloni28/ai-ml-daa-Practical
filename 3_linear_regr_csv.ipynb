```python
import pandas as pd

data = pd.read_csv("data1.csv")

X = data['X'].tolist()
Y = data['Y'].tolist()

n = len(X)
```

```python
sum_x = sum(X)
sum_y = sum(Y)

sum_xy = sum([X[i]*Y[i] for i in range(n)])
sum_x2 = sum([x*x for x in X])

print(sum_x, sum_y, sum_xy, sum_x2)
```

```python
m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)
print("Slope (m):", m)
```

```python
c = (sum_y - m*sum_x) / n
print("Intercept (c):", c)
```

```python
print("Equation of best fit line:")
print(f"y = {m:.2f}x + {c:.2f}")
```

```python
import matplotlib.pyplot as plt

y_pred = [m*x + c for x in X]

plt.scatter(X, Y)
plt.plot(X, y_pred)

plt.xlabel("Hours Driving")
plt.ylabel("Risk Score")
plt.title("Linear Regression")

plt.show()
```
